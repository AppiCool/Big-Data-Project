```
from pyspark.sql.functions import col, lag
from pyspark.sql.window import Window
from sklearn.preprocessing import MinMaxScaler
import numpy as np

def derive_temporal(temporal_features_spark_df, shift_values):
    for shift_value in shift_values:
        name = 'High_Lag_{}'.format(shift_value)
        temporal_features_spark_df = temporal_features_spark_df.withColumn(name, lag("High", shift_value).over(Window.orderBy("Date")))

    temporal_features_spark_df = temporal_features_spark_df.dropna()
    return temporal_features_spark_df

def scale_to_numpy(scale_to_numpy_spark_stock_df):
    minMaxScaler = MinMaxScaler()
    scale_to_numpy_spark_stock_df = scale_to_numpy_spark_stock_df.toPandas()
    
    scale_to_numpy_spark_stock_df.iloc[:, 1:] = minMaxScaler.fit_transform(scale_to_numpy_spark_stock_df.iloc[:, 1:])
    
    scale_to_numpy_spark_stock_np = scale_to_numpy_spark_stock_df.values
    return scale_to_numpy_spark_stock_np

num_shifts = 10

scale_to_numpy_array = scale_to_numpy(derive_temporal(stockInfo, range(1, num_shifts + 1)))
```

```
def divide_dataset(stock_numpy):
    target_index = None
    feature_indices = []
    
    for i in range(num_shifts + 2):
        if i == 1:
            target_index = i
        else:
            feature_indices.append(i)

    print(feature_indices)
    input_features = stock_numpy[:, feature_indices]
    target_values = stock_numpy[:, [target_index]] if target_index is not None else None

    return input_features, target_values

exchange_rate_data = {
    'GBPtoEGP': divide_dataset(scale_to_numpy_array)
}

X_GBPtoEGP, Y_GBPtoEGP = exchange_rate_data['GBPtoEGP']
```

```
import numpy as np

def modify_and_reshape(train_data_part, test_data_part):
    modified_tr_data = train_data_part[:, 1:]
    modified_te_data = test_data_part[:, 1:] 
    
    reshaped_tr_data = modified_tr_data.reshape(modified_tr_data.shape[0], modified_tr_data.shape[1], 1)
    reshaped_te_data = modified_te_data.reshape(modified_te_data.shape[0], modified_te_data.shape[1], 1)
    
    return reshaped_tr_data, reshaped_te_data

X_train_modified, X_test_modified = modify_and_reshape(X_GBPtoEGP_train, X_GBPtoEGP_test)
```

```
def quadratic_average(verified_data, forcasted_numericals):
    squared_delta = list()
    if type(verified_data) is not list:
        verified_data = list(verified_data)
    for i in range(len(verified_data)):
        squared_delta.append((verified_data[i] - forcasted_numericals[i]) ** 2)
    return np.sqrt((sum(squared_delta) / len(verified_data)).astype(np.float32))

def estimated_quantity(verified_data, forcasted_numericals):
    verified_data = list(verified_data)
    squared_diff_list = []
    for cur_i in range(len(verified_data)):
        squared_diff_list.append((verified_data[cur_i] - forcasted_numericals[cur_i]) ** 2)
    return 1 - (sum(squared_diff_list) / len(verified_data))
```

```
class LSTM:
    
    validations = []
    def __init__(self, component_count, element_reference, adaptation_factor):
        self.validations = []
        self.component_count, self.element_reference, self.adaptation_factor = component_count, element_reference, adaptation_factor

        self.kp_front, self.kp_internal, self.kp_ingestion, self.st_cell, self.kp_external, self.kp_tuf = [np.zeros((element_reference, component_count, 1)) for _ in range(6)]

        slim_line, solid_line = -(1.0 / np.sqrt(element_reference)), (1.0 / np.sqrt(element_reference))
        self.kp_foreline, self.kp_ingest, self.kp_box, self.kp_linear = [slim_line + np.random.random((component_count, 1 + component_count)) * (solid_line - slim_line) for _ in range(4)]
        self.kp_forward_1, self.kp_internal_1, self.kp_box_1, self.kp_exp_1 = [slim_line + np.random.random((component_count, 1)) * (solid_line - slim_line) for _ in range(4)]
        self.kp_internal_vec, self.kp_bilinear_prob = slim_line + np.random.random((1, component_count)) * (solid_line - slim_line), slim_line + np.random.random((1, 1)) * (solid_line - slim_line)
            
    def progression(self, input, assist_data):
        forget_gate = self.contoured(np.dot(self.kp_foreline, assist_data) + self.kp_forward_1)
        input_gate = self.contoured(np.dot(self.kp_ingest, assist_data) + self.kp_internal_1)
        output_gate = self.contoured(np.dot(self.kp_linear, assist_data) + self.kp_exp_1)
        candidate_state = self.hyperbolic_fn(np.dot(self.kp_box, assist_data) + self.kp_box_1)
        internal_state = self.st_cell[-1] * forget_gate + input_gate * candidate_state
        hidden_state = self.hyperbolic_fn(internal_state) * output_gate
        return forget_gate, input_gate, output_gate, candidate_state, internal_state, hidden_state

    def forward_propagation_func(self, input):
        for given_box in range(1, self.element_reference):
            assist_data = np.vstack((self.kp_tuf[given_box - 1], input[given_box]))
            forget_gate, input_gate, output_gate, candidate_state, internal_state, hidden_state = self.progression(input[given_box], assist_data)
            self.kp_front[given_box] = forget_gate
            self.kp_internal[given_box] = input_gate
            self.kp_ingestion[given_box] = candidate_state
            self.st_cell[given_box] = internal_state     
            self.kp_external[given_box] = output_gate
            self.kp_tuf[given_box] = hidden_state
        predcition_Y = np.dot(self.kp_internal_vec, self.kp_tuf[-1]) + self.kp_bilinear_prob
        return predcition_Y
    

    def start_diff_array(self,nodal_unit, count):
        return [np.zeros((nodal_unit, 1)) for _ in range(count)]

    def bck_prpg_fun(self, Y_true, Y_pred,input):
        kp_forerun_disparity = self.start_diff_array(self.component_count, self.element_reference)
        kp_imbibe_disparity = self.start_diff_array(self.component_count, self.element_reference)
        kp_hold_disparity = self.start_diff_array(self.component_count, self.element_reference)
        kp_box_disparity = self.start_diff_array(self.component_count, self.element_reference)
        kp_efficient_disparity = self.start_diff_array(self.component_count, self.element_reference)
        kp_hold_disparity_diff = self.start_diff_array(self.component_count, self.element_reference)

        lop_weight_include = lop_weight_ingest = lop_weight_push = lop_weight_box = np.zeros((self.component_count, 1 + self.component_count))
        lop_dual_optimal = lop_dual_ingest = lop_dual_front = np.zeros((self.component_count, 1))
        lop_dual_box = np.zeros_like(self.kp_box_1)

        tuy_change_auth = Y_true - Y_pred
        lop_dual_convert = tuy_change_auth
        lop_dual_end = np.dot(tuy_change_auth, self.kp_tuf[-1].T)
        

        for active_position in reversed(range(self.element_reference-1)):

            kp_hold_disparity_diff[active_position] = self.kp_internal_vec.T @ tuy_change_auth + kp_hold_disparity_diff[active_position+1]
            kp_box_disparity[active_position] =  kp_hold_disparity_diff[active_position] * self.kp_external[active_position]  * 1-(self.hyperbolic_fn(self.st_cell[active_position]))**2 + kp_box_disparity[active_position+1]*self.kp_front[active_position+1]
            kp_hold_disparity[active_position] = kp_box_disparity[active_position] * self.kp_internal[active_position]  * 1-(self.kp_ingestion[active_position])**2
            kp_imbibe_disparity[active_position] = kp_box_disparity[active_position] *self.kp_ingestion[active_position] * self.kp_internal[active_position] * (1- self.kp_internal[active_position])
            kp_forerun_disparity[active_position] = kp_box_disparity[active_position]  * self.st_cell[active_position-1] * self.kp_front[active_position]*(1-self.kp_front[active_position])
            kp_efficient_disparity[active_position] = kp_hold_disparity_diff[active_position] * self.hyperbolic_fn(self.st_cell[active_position]) * self.kp_external[active_position]*(1-self.kp_external[active_position])
            input_vector = np.concatenate((self.kp_tuf[active_position-1], input[active_position].reshape(-1,1)),axis=0)
            lop_weight_push =lop_weight_push + kp_forerun_disparity[active_position] @ input_vector.T
            lop_weight_ingest = lop_weight_ingest + np.matmul(kp_imbibe_disparity[active_position] , input_vector.T)
            lop_weight_box = lop_weight_box + np.matmul(kp_box_disparity[active_position] ,input_vector.T)
            lop_weight_include = lop_weight_include + np.matmul(kp_efficient_disparity[active_position], input_vector.T)
            lop_dual_ingest = lop_dual_ingest + kp_imbibe_disparity[active_position]
            lop_dual_front = lop_dual_front + kp_forerun_disparity[active_position]
            lop_dual_optimal = lop_dual_optimal + kp_efficient_disparity[active_position]
            lop_dual_box = lop_dual_box + kp_box_disparity[active_position]            

        return lop_dual_end, lop_dual_convert, lop_weight_push, lop_dual_front, lop_weight_ingest, lop_dual_ingest, lop_weight_include, lop_dual_optimal, lop_weight_box, lop_dual_box  
      
    def fit(self, tabulate, input, original_dual):
        output_vector = np.mean(np.array(original_dual))
        for counter in range(tabulate):
            yu_io_tabulate = 0 
            kp_array=list()
            rkp_array=list()
            array_dual_1=list()
            for pointer, input_vector in enumerate(input):
                Y_pred = self.forward_propagation_func(input_vector)
                array_dual_1.append(Y_pred)
                lop_dual_end, lop_dual_convert, lop_weight_push, lop_dual_front, lop_weight_ingest, lop_dual_ingest, lop_weight_include, lop_dual_optimal, lop_weight_box, lop_dual_box = self.bck_prpg_fun(original_dual[pointer], Y_pred, input_vector)

                conv_lop_forward = np.mean(lop_weight_push, axis=0) / self.element_reference
                conv_lop_ingest = np.mean(lop_weight_ingest, axis=0) / self.element_reference
                conv_lop_box = np.mean(lop_weight_box, axis=0) / self.element_reference
                conv_lop_ext = np.mean(lop_weight_include, axis=0) / self.element_reference
                conv_lop_dual_split = np.mean(lop_dual_front, axis=0) / self.element_reference
                conv_lop_dual_ingest = np.mean(lop_dual_ingest, axis=0) / self.element_reference
                conv_lop_dual_box = np.mean(lop_dual_box, axis=0) / self.element_reference
                conv_lop_dual_excced = np.mean(lop_dual_optimal, axis=0) / self.element_reference

                self.kp_foreline = self.kp_foreline + self.adaptation_factor * conv_lop_forward
                self.kp_ingest = self.kp_ingest + self.adaptation_factor * conv_lop_ingest
                self.kp_box = self.kp_box + self.adaptation_factor * conv_lop_box
                self.kp_linear = self.kp_linear + self.adaptation_factor * conv_lop_ext
                self.kp_forward_1 = self.kp_forward_1 + self.adaptation_factor * conv_lop_dual_split
                self.kp_internal_1 = self.kp_internal_1 + self.adaptation_factor * conv_lop_dual_ingest
                self.kp_box_1 = self.kp_box_1 + self.adaptation_factor * conv_lop_dual_box
                self.kp_exp_1 = self.kp_exp_1 + self.adaptation_factor * conv_lop_dual_excced
                self.kp_bilinear_prob = self.kp_bilinear_prob + self.adaptation_factor * lop_dual_convert
                self.kp_internal_vec = self.kp_internal_vec + self.adaptation_factor * lop_dual_end

                yu_io_tabulate += ((original_dual[pointer] - Y_pred) ** 2) / 2
                kp_array.append((original_dual[pointer] - Y_pred) ** 2)
                rkp_array.append((Y_pred - output_vector) ** 2)
            
            
            accuracy= estimated_quantity(original_dual,array_dual_1)
            quadratic_inconsistency=quadratic_average(original_dual,array_dual_1)
            print("counter: " ,counter, " Training deficit: ", (yu_io_tabulate/len(input))[0][0], " points: ", accuracy," quadratic_inconsistency: ", quadratic_inconsistency)
            self.validations.append((counter,(yu_io_tabulate/len(input))[0][0],accuracy[0][0],quadratic_inconsistency[0][0]))
    
    def forecast(self, ingest_k):         
        return [self.forward_propagation_func(k) for k in ingest_k]
    
    def contoured(self, ingest_first):
        return 1 / (1 + np.exp(-ingest_first.astype(np.float32)))
    
    def hyperbolic_fn(self, ingest_first):
        return np.tanh(ingest_first.astype(np.float32))
```

```
def custom_LSTM(input_data, target_train_data, input_test_data, target_test_data):
    input_neurons = 10
    output_neurons = 1
    num_neurons = 15
    num_epochs = 10
    learning_rate = 0.05

    my_custom_LSTM = LSTM(num_neurons, input_neurons, learning_rate)
    my_custom_LSTM.fit(num_epochs, input_data, target_train_data)

    predicted_training = my_custom_LSTM.forecast(input_data)
    predicted_test = my_custom_LSTM.forecast(input_test_data)

    predicted_training = np.concatenate(predicted_training)
    predicted_test = np.concatenate(predicted_test)

    r2_train = estimated_quantity(list(target_train_data), list(predicted_training))[0]
    rmse_train = quadratic_average(list(target_train_data), list(predicted_training))[0]
    r2_test = estimated_quantity(list(target_test_data), list(predicted_test))[0]
    rmse_test = quadratic_average(list(target_test_data), list(predicted_test))[0]

    print("The R2 score for the training set is", r2_train)
    print("The RMSE score for the training set is", rmse_train)
    print("The R2 score of the test set is", r2_test)
    print("The RMSE score of the test set is", rmse_test)

    return predicted_test, my_custom_LSTM.validations
```

```
import pandas as pd
import matplotlib.pyplot as custom_diagram

def diagram_predictions(actual_dates,actual_high_prices,predicted_high_prices,plot_title):
    dates=actual_dates[:,0]
    df_plot=pd.DataFrame(np.concatenate((dates.reshape(-1,1),actual_high_prices.reshape(-1,1),predicted_high_prices),axis=1),columns=["Date", "Actual_High_Price", "Predicted_High_Price"])
    custom_diagram.figure(figsize=(18, 9))
    df_plot.set_index('Date', inplace=True)
    custom_diagram.scatter(df_plot.index, df_plot["Actual_High_Price"], color="brown", label='Original Exchange Price')
    custom_diagram.scatter(df_plot.index, df_plot["Predicted_High_Price"], color="red", label='Forecasted Exchange Price')
    custom_diagram.xlabel('Yearwise Progression')
    custom_diagram.ylabel('Exchange Price')
    custom_diagram.legend()
    custom_diagram.title(plot_title)
    custom_diagram.show()
```
